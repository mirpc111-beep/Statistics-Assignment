# (i) Write a Python function to calculate the mean of a given list of numbers. Create a function to find the median of a list of numbers?

````python
import random

# Generate the list of 100 integers (90 to 130)
int_list = [random.randint(90, 130) for _ in range(100)]
print("Generated List:", int_list)

# Function to calculate the mean
def mean(values):
    return sum(values) / len(values)

# Function to calculate the median
def median(values):
    values_sorted = sorted(values)
    n = len(values_sorted)
    mid = n // 2

    if n % 2 == 0:   # even count
        return (values_sorted[mid - 1] + values_sorted[mid]) / 2
    else:            # odd count
        return values_sorted[mid]

# Print results
print("Mean:", mean(int_list))
print("Median:", median(int_list))
````

# (ii) Develop a program to compute the mode of a list of integers?

````python
import random
from collections import Counter

# Generate the list of 100 integers (90 to 130)
int_list = [random.randint(90, 130) for _ in range(100)]
print("Generated List:", int_list)

# Function to compute the mode
def mode(values):
    freq = Counter(values)               # Count occurrences of each number
    max_count = max(freq.values())       # Find highest frequency
    modes = [num for num, count in freq.items() if count == max_count]
    return modes                         # Return list in case of multiple modes

# Print the mode
print("Mode:", mode(int_list))
````

# (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights?

````python
import random

# Generate the list of 100 integers (90 to 130)
int_list = [random.randint(90, 130) for _ in range(100)]
print("Generated List:", int_list)

# Function to calculate weighted mean
def weighted_mean(values, weights):
    if len(values) != len(weights):
        raise ValueError("Values and weights must have the same length.")

    return sum(v * w for v, w in zip(values, weights)) / sum(weights)

# Example: create 100 weights (all 1s or random weights)
weights = [random.randint(1, 5) for _ in range(100)]

# Calculate weighted mean
print("Weighted Mean:", weighted_mean(int_list, weights))
````

# (iv) Write a Python function to find the geometric mean of a list of positive numbers?

````python
import random
import math

# Step 1: Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]
print("Values:", int_list)

# Step 2: Function to calculate geometric mean
def geometric_mean(values):
    product = 1
    n = len(values)

    for v in values:
        product *= v   # multiply all values together

    return product ** (1 / n)   # take nth root

# Step 3: Print geometric mean
print("Geometric Mean:", geometric_mean(int_list))
````

# (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of outliers?

````python
import random

# Step 1: Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]
print("Original List:", int_list)

# Step 2: Function to calculate trimmed mean
def trimmed_mean(data, percentage):
    """
    Calculates the trimmed mean of a list by excluding
    a certain percentage of the lowest and highest values.

    Parameters:
    data (list): List of numbers
    percentage (float): Percentage of values to trim from each end (0-50)

    Returns:
    float: Trimmed mean
    """
    if not 0 <= percentage <= 50:
        raise ValueError("Percentage must be between 0 and 50")

    n = len(data)
    trim_count = int(n * percentage / 100)

    # Sort the data
    sorted_data = sorted(data)

    # Trim the lowest and highest values
    trimmed_data = sorted_data[trim_count : n - trim_count]

    # Calculate the mean of trimmed data
    return sum(trimmed_data) / len(trimmed_data)

percentage_to_trim = 10  # trimming 10% from both ends
result = trimmed_mean(int_list, percentage_to_trim)
print(f"Trimmed Mean (trimming {percentage_to_trim}% of outliers):", result)
````

# (iv) Implement a function to compute the interquartile range (IQR) of a list of values?

````python
import random
import numpy as np

# Step 1: Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]
print("First 20 numbers in the list:", int_list2[:20])

# Step 2: Function to calculate IQR
def calculate_iqr(data):
    """
    Calculates the interquartile range (IQR) of a list of numbers.

    Parameters:
    data (list): List of numerical values

    Returns:
    float: Interquartile range (Q3 - Q1)
    """
    if not data:
        return None  # Return None for empty list

    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    return Q3 - Q1

iqr_value = calculate_iqr(int_list2)
print("Interquartile Range (IQR) of the list:", iqr_value)
````

# (v) Build a program to calculate the coefficient of variation for a dataset?

````python
import random
import statistics

# Step 1: Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]
print("First 20 numbers in the list:", int_list2[:20])

# Step 2: Function to calculate coefficient of variation
def coefficient_of_variation(data):
    """
    Calculates the coefficient of variation (CV) of a dataset.

    Parameters:
    data (list): List of numerical values

    Returns:
    float: CV in percentage
    """
    if not data:
        return None  # Return None for empty list

    mean = statistics.mean(data)
    if mean == 0:
        return None  # Avoid division by zero

    std_dev = statistics.stdev(data)
    cv = (std_dev / mean) * 100
    return cv

cv_value = coefficient_of_variation(int_list2)
print("Coefficient of Variation (CV) of the list: {:.2f}%".format(cv_value))
````

# (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers?

````python
import random
import statistics

# Step 1: Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]
print("First 20 numbers in the list:", int_list2[:20])

# Step 2: Function to calculate Mean Absolute Deviation
def mean_absolute_deviation(data):
    """
    Calculates the Mean Absolute Deviation (MAD) of a dataset.

    Parameters:
    data (list): List of numerical values

    Returns:
    float: MAD of the list
    """
    if not data:
        return None  # Return None for empty list

    mean_value = statistics.mean(data)
    mad = sum(abs(x - mean_value) for x in data) / len(data)
    return mad

mad_value = mean_absolute_deviation(int_list2)
print("Mean Absolute Deviation (MAD) of the list:", mad_value)
````

# (vii) Create a program to calculate the quartile deviation of a list of values?

````python
import random
import numpy as np

# Step 1: Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]
print("First 20 numbers in the list:", int_list2[:20])

# Step 2: Function to calculate Quartile Deviation
def quartile_deviation(data):
    """
    Calculates the Quartile Deviation (Semi-Interquartile Range) of a dataset.

    Parameters:
    data (list): List of numerical values

    Returns:
    float: Quartile Deviation
    """
    if not data:
        return None  # Return None for empty list

    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    return (Q3 - Q1) / 2

qd_value = quartile_deviation(int_list2)
print("Quartile Deviation (QD) of the list:", qd_value)
````

# (viii) Implement a function to find the range-based coefficient of dispersion for a dataset?

````python
import random
import numpy as np

# Step 1: Generate a list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]
print("First 20 numbers in the list:", int_list2[:20])

# Step 2: Function to calculate range-based coefficient of dispersion
def range_based_cod(data):
    """
    Calculates the range-based coefficient of dispersion (COD) of a dataset.

    Parameters:
    data (list): List of numerical values

    Returns:
    float: Coefficient of Dispersion
    """
    if not data:
        return None  # Return None for empty list

    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    if Q3 + Q1 == 0:
        return None  # Avoid division by zero

    cod = (Q3 - Q1) / (Q3 + Q1)
    return cod

cod_value = range_based_cod(int_list2)
print("Range-based Coefficient of Dispersion (COD) of the list:", cod_value)
````

# 3)  Write a Python class representing a discrete random variable with methods to calculate its expected value and variance?

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """
        Initializes the discrete random variable.

        Parameters:
        values (list): Possible outcomes of the random variable
        probabilities (list): Corresponding probabilities of each outcome
        """
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must have the same length.")
        if not abs(sum(probabilities) - 1) < 1e-6:
            raise ValueError("Probabilities must sum to 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """
        Calculates the expected value (mean) of the random variable.
        """
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self):
        """
        Calculates the variance of the random variable.
        """
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for v, p in zip(self.values, self.probabilities))

values = [0, 1, 2, 3]  # Possible outcomes
probabilities = [0.1, 0.2, 0.4, 0.3]  # Probabilities corresponding to each outcome

rv = DiscreteRandomVariable(values, probabilities)

print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
````

# 4) Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes?

# 5) Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance?

````python
import numpy as np

def sample_distribution(dist_name, params, sample_size=1000):
    """
    Generates random samples from a specified probability distribution
    and calculates their mean and variance.

    Parameters:
    dist_name (str): Name of the distribution ('binomial' or 'poisson')
    params (dict): Parameters for the distribution
                   - Binomial: {'n': trials, 'p': probability}
                   - Poisson: {'lam': lambda}
    sample_size (int): Number of samples to generate

    Returns:
    tuple: (samples, mean, variance)
    """
    if dist_name.lower() == 'binomial':
        n = params.get('n')
        p = params.get('p')
        samples = np.random.binomial(n, p, sample_size)

    elif dist_name.lower() == 'poisson':
        lam = params.get('lam')
        samples = np.random.poisson(lam, sample_size)

    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples, ddof=1)  # sample variance
    return samples, mean, variance


# Binomial distribution: 10 trials, probability 0.5
binom_samples, binom_mean, binom_var = sample_distribution('binomial', {'n': 10, 'p': 0.5}, 1000)
print("Binomial Distribution:")
print("Mean:", binom_mean)
print("Variance:", binom_var)

# Poisson distribution: lambda = 4
poisson_samples, poisson_mean, poisson_var = sample_distribution('poisson', {'lam': 4}, 1000)
print("\nPoisson Distribution:")
print("Mean:", poisson_mean)
print("Variance:", poisson_var)
````

# (iii) Write a function that calculates the covariance between two column?

````python
import seaborn as sns

# Step 1: Load the tips dataset
tips = sns.load_dataset('tips')

# Step 2: Function to calculate covariance
def calculate_covariance(column1, column2):
    """
    Calculates the covariance between two columns.

    Parameters:
    column1, column2 (pd.Series or list): Two numerical columns

    Returns:
    float: Covariance value
    """
    mean1 = column1.mean()
    mean2 = column2.mean()
    n = len(column1)

    covariance = sum((x - mean1) * (y - mean2) for x, y in zip(column1, column2)) / (n - 1)
    return covariance

cov_total_tip = calculate_covariance(tips['total_bill'], tips['tip'])
print("Covariance between total_bill and tip:", cov_total_tip)
````

# (iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns?

````python
import seaborn as sns
import numpy as np

# Step 1: Load the tips dataset
tips = sns.load_dataset('tips')

# Step 2: Function to calculate Pearson correlation coefficient
def pearson_correlation(column1, column2):
    """
    Calculates the Pearson correlation coefficient between two columns.

    Parameters:
    column1, column2 (pd.Series or list): Two numerical columns

    Returns:
    float: Pearson correlation coefficient
    """
    mean1 = np.mean(column1)
    mean2 = np.mean(column2)
    numerator = sum((x - mean1) * (y - mean2) for x, y in zip(column1, column2))
    denominator = ((sum((x - mean1)**2 for x in column1) * sum((y - mean2)**2 for y in column2)) ** 0.5)

    return numerator / denominator

pearson_corr = pearson_correlation(tips['total_bill'], tips['tip'])
print("Pearson correlation coefficient between total_bill and tip:", pearson_corr)
````

# 8) Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution?

````python
import math

def normal_pdf(x, mu=0, sigma=1):
    """
    Calculates the probability density function (PDF) of a normal distribution.

    Parameters:
    x (float or list of floats): Value(s) at which to evaluate the PDF
    mu (float): Mean of the normal distribution
    sigma (float): Standard deviation of the normal distribution

    Returns:
    float or list of floats: PDF value(s) at x
    """
    def pdf_single(val):
        return (1 / (sigma * math.sqrt(2 * math.pi))) * math.exp(-((val - mu)**2) / (2 * sigma**2))

    # Support for single value or list of values
    if isinstance(x, (list, tuple)):
        return [pdf_single(val) for val in x]
    else:
        return pdf_single(x)

x_values = [45, 50, 55]
pdf_values = normal_pdf(x_values, mu=50, sigma=5)
print("PDF values at", x_values, ":", pdf_values)
````

# 9) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution?

````python
import math

def exponential_cdf(x, lam):
    """
    Calculates the CDF of an exponential distribution.

    Parameters:
    x (float or list of floats): Value(s) at which to evaluate the CDF
    lam (float): Rate parameter (lambda > 0)

    Returns:
    float or list of floats: CDF value(s) at x
    """
    def cdf_single(val):
        if val < 0:
            return 0  # CDF = 0 for x < 0
        return 1 - math.exp(-lam * val)

    if isinstance(x, (list, tuple)):
        return [cdf_single(val) for val in x]
    else:
        return cdf_single(x)

x_values = [0, 1, 2, 3]
lam = 0.5  # Rate parameter
cdf_values = exponential_cdf(x_values, lam)
print("CDF values at", x_values, ":", cdf_values)
````

# 10) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution?

````python
import math

def poisson_pmf(k, lam):
    """
    Calculates the PMF of a Poisson distribution.

    Parameters:
    k (int or list of ints): Number(s) of events
    lam (float): Rate parameter (lambda > 0)

    Returns:
    float or list of floats: PMF value(s) at k
    """
    def pmf_single(val):
        if val < 0 or not isinstance(val, int):
            return 0  # PMF is 0 for negative or non-integer events
        return (lam ** val * math.exp(-lam)) / math.factorial(val)

    # Support single value or list of values
    if isinstance(k, (list, tuple)):
        return [pmf_single(val) for val in k]
    else:
        return pmf_single(k)

k_values = [0, 1, 2, 3, 4]
lam = 2.5
pmf_values = poisson_pmf(k_values, lam)
print("PMF values at", k_values, ":", pmf_values)
````

```python
 70 purchases out of 1000 visitors

  ```

# Apply z-test to find which layout is successful

````python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Step 1: Generate the data
old_layout = np.array([1]*50 + [0]*950)   # 50 purchases out of 1000
new_layout = np.array([1]*70 + [0]*930)   # 70 purchases out of 1000

# Step 2: Count successes and total observations
successes = np.array([old_layout.sum(), new_layout.sum()])
n_obs = np.array([len(old_layout), len(new_layout)])

# Step 3: Perform two-proportion z-test
z_stat, p_value = proportions_ztest(successes, n_obs)

print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: New layout has a significantly higher conversion rate.")
else:
    print("Fail to reject null hypothesis: No significant difference between layouts.")
````

# 15. A company is testing two different website layouts to see which one leads to higher click-through rates. Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.


Use the following data:

python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

````python
import numpy as np
from scipy.stats import ttest_ind

def ab_test_ttest(group_a, group_b):
    """
    Performs an A/B test using an independent two-sample t-test.

    Returns:
      t_stat: t-statistic
      df: degrees of freedom
      p_value: p-value
    """
    # Convert to numpy arrays
    a = np.array(group_a)
    b = np.array(group_b)

    # Sample sizes
    n1 = len(a)
    n2 = len(b)

    # Sample means
    mean1 = np.mean(a)
    mean2 = np.mean(b)

    # Sample variances
    var1 = np.var(a, ddof=1)
    var2 = np.var(b, ddof=1)

    # t-statistic (Welchs formula)
    t_stat = (mean1 - mean2) / np.sqrt(var1/n1 + var2/n2)

    # Degrees of freedom (WelchSatterthwaite equation)
    df = (var1/n1 + var2/n2)**2 / (
        (var1**2) / (n1**2 * (n1 - 1)) +
        (var2**2) / (n2**2 * (n2 - 1))
    )

    # p-value (two-tailed)
    # Using scipy for p-value of computed t-stat
    from scipy.stats import t
    p_value = 2 * t.sf(abs(t_stat), df)

    return t_stat, df, p_value


# ----------------------------------------------------------
# Given Data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Run A/B test
t_stat, df, p_value = ab_test_ttest(layout_a_clicks, layout_b_clicks)

print("T-statistic:", t_stat)
print("Degrees of freedom:", df)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Significant difference. One layout performs better.")
else:
    print("Result: No significant difference detected.")
````

# 17. A school district introduces an educational intervention program to improve math scores. Write a Python function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to determine if the intervention had a significant impact.


Use the following data of test score:


python

  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]

  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

````python
import numpy as np
from scipy.stats import t

def intervention_effect(pre_scores, post_scores):
    """
    Performs a paired t-test to evaluate intervention effect.
    Returns t-statistic, df, and p-value.
    """
    pre = np.array(pre_scores)
    post = np.array(post_scores)

    # Differences (post - pre)
    diff = post - pre

    # Sample size
    n = len(diff)

    # Mean and std of differences
    mean_diff = np.mean(diff)
    std_diff = np.std(diff, ddof=1)

    # t-statistic for paired t-test
    t_stat = mean_diff / (std_diff / np.sqrt(n))

    # Degrees of freedom
    df = n - 1

    # Two-tailed p-value
    p_val = 2 * t.sf(abs(t_stat), df)

    return t_stat, df, p_val


# ----------------------------------------------------
# Given Data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Run test
t_stat, df, p_value = intervention_effect(pre_intervention_scores, post_intervention_scores)

print("T-statistic:", t_stat)
print("Degrees of freedom:", df)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Significant improvement  the intervention worked.")
else:
    print("Result: No statistically significant improvement detected.")
````

# 18. An HR department wants to investigate if there's a gender-based salary gap within the company. Develop a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically significant difference between the average salaries of male and female employees.


Use the below code to generate synthetic data:


python

# Generate synthetic salary data for male and female employees

np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)

female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

````python
import numpy as np
from scipy.stats import t

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility
male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Function to perform independent t-test (Welch's t-test)
def salary_gap_ttest(male, female):
    n1, n2 = len(male), len(female)
    mean1, mean2 = np.mean(male), np.mean(female)
    var1, var2 = np.var(male, ddof=1), np.var(female, ddof=1)

    # Standard error
    se = np.sqrt(var1/n1 + var2/n2)

    # t-statistic
    t_stat = (mean1 - mean2) / se

    # Degrees of freedom (Welch-Satterthwaite equation)
    df = (var1/n1 + var2/n2)**2 / ((var1**2)/((n1**2)*(n1-1)) + (var2**2)/((n2**2)*(n2-1)))

    # Two-tailed p-value
    p_value = 2 * (1 - t.cdf(abs(t_stat), df))

    return t_stat, df, p_value

# Run the test
t_stat, df, p_value = salary_gap_ttest(male_salaries, female_salaries)

# Display results
print("Male Salaries Mean :", np.mean(male_salaries))
print("Female Salaries Mean :", np.mean(female_salaries))
print("\nT-statistic:", t_stat)
print("Degrees of freedom:", df)
print("P-value:", p_value)

if p_value < 0.05:
    print("\nConclusion: There IS a statistically significant difference in salaries.")
else:
    print("\nConclusion: There is NO statistically significant difference in salaries.")
````

python

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate synthetic data
np.random.seed(0)
age_groups = np.random.choice(['1830', '3150', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create DataFrame
df = pd.DataFrame({
    'Age Group': age_groups,
    'Preference': voter_preferences
})

# Create contingency table
contingency_table = pd.crosstab(df['Age Group'], df['Preference'])
print("Contingency Table:")
print(contingency_table)

# Perform Chi-Square Test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)
print("\nExpected Frequencies:")
print(expected)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: YES  There is a significant association between age group and voter preference.")
else:
    print("\nConclusion: NO  There is NO significant association between age group and voter preference.")
````
